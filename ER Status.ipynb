# **Splitting Independent variable and Dependent variable**

```python
x=data.drop('ER Status',axis=1)
y=data['ER Status']
```

```python
x.dtypes
```

```python
x.dtypes.value_counts()
```

```python
y.value_counts()
```

```python
y
```

# **Label Encoding**

```python
from sklearn.preprocessing import LabelEncoder
```

```python
x=x.apply(LabelEncoder().fit_transform)
y=LabelEncoder().fit_transform(y)
```

```python
y
```

```python
x.dtypes
```

# **Feature Scale**

```python
from sklearn.preprocessing import StandardScaler
x=StandardScaler().fit_transform(x)
```

# **Splitting Traing and Testing Dataset**

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 0.3, random_state=42)
```

# **Classification Model**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score,classification_report, confusion_matrix
```

# **Logistic Regression**

```python
l1=LogisticRegression()
l1.fit(x_train,y_train)
y_pred=l1.predict(x_test)
Accuracy_lr=accuracy_score(y_test,y_pred)
Accuracy_lr*100
```

```python
import seaborn as sns
cm = confusion_matrix(y_test, y_pred)
#print(cm)
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test,y_pred))
```

# **KNN Classifier**

```python
knn=KNeighborsClassifier(n_neighbors=5,metric='minkowski',p=1)
knn.fit(x_train,y_train)
knn_pred=knn.predict(x_test)
Accuracy_knn=accuracy_score(y_test,knn_pred)
Accuracy_knn*100
```

```python
cm = confusion_matrix(y_test, knn_pred)
#print(cm)
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test,knn_pred))
```

# **Random Forest classifier**

```python
rfc=RandomForestClassifier()
rfc.fit(x_train,y_train)
rfc_pred=rfc.predict(x_test)
Accuracy_rf=accuracy_score(y_test,rfc_pred)
Accuracy_rf*100
```

```python
cm = confusion_matrix(y_test,rfc_pred)
#print(cm)
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test,rfc_pred))
```

# **ROC curve**

```python
from sklearn.metrics import roc_curve
fpr1, tpr1, thresh1 = roc_curve(y_test, rfc_pred, pos_label=1)

# roc curve for tpr = fpr
random_probs = [0 for i in range(len(y_test))]
p_fpr, p_tpr, _ = roc_curve(y_test, random_probs, pos_label=1)
```

```python
from sklearn.metrics import roc_auc_score

# auc scores
auc_score1 = roc_auc_score(y_test, rfc_pred)
print(auc_score1)
```

# **AUC score**

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn')

# plot roc curves
plt.plot(fpr1, tpr1, linestyle='--',color='orange', label='Random Forest Classifier')
plt.plot(p_fpr, p_tpr, linestyle='--', color='blue')
# title
plt.title('ROC curve')
# x label
plt.xlabel('False Positive Rate')
# y label
plt.ylabel('True Positive rate')

plt.legend(loc='best')
plt.savefig('ROC',dpi=300)
plt.show();
```

# **Decision Tree Classifier**

```python
dt=DecisionTreeClassifier(criterion = "entropy",random_state=20)
dt.fit(x_train,y_train)
dt_pred=dt.predict(x_test)
Accuracy_dt=accuracy_score(y_test,dt_pred)
Accuracy_dt*100
```

```python
cm = confusion_matrix(y_test, dt_pred)
#print(cm)
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test,dt_pred))
```

# **Support Vector Machine**

```python
svm = SVC(kernel="rbf", gamma=0.5, C=1.0)
svm.fit(x_train,y_train)
svm_pred=svm.predict(x_test)
Accuracy_svm=accuracy_score(y_test,svm_pred)
Accuracy_svm*100
```

```python
cm = confusion_matrix(y_test,svm_pred)
#print(cm)
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test,svm_pred))
```

# **Naive Bayes**

```python
gnb = GaussianNB()
gnb.fit(x_train,y_train)
gnb_pred=gnb.predict(x_test)
Accuracy_nb=accuracy_score(y_test,gnb_pred)
Accuracy_nb*100
```

```python
cm = confusion_matrix(y_test,gnb_pred)
#print(cm)
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test,gnb_pred))
```
