# **Splitting Independent variable and Dependent variable**

```python
x=data.drop("ER status measured by IHC",axis=1)
y=data["ER status measured by IHC"]
```

# **Label Encoding**

```python
from sklearn.preprocessing import LabelEncoder
```

```python
x=x.apply(LabelEncoder().fit_transform)
y=LabelEncoder().fit_transform(y)
```

```python
y
```

```python
x.dtypes
```

# **Feature Scale**

```python
from sklearn.preprocessing import StandardScaler
x=StandardScaler().fit_transform(x)
```

# **K flod Cross Validation**

```python
from sklearn.model_selection import KFold, cross_val_score
k_folds = KFold(n_splits = 5)
```

# **Classification Model**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score,classification_report, confusion_matrix
```

# **Logistic Regression**

```python
l1=LogisticRegression()
scores = cross_val_score(l1, x, y, cv = k_folds)
LR=scores.mean()

print("Cross Validation Scores: ", scores)
print("Average CV Score: ", scores.mean())
```

# **KNN Classifier**

```python
knn=KNeighborsClassifier(n_neighbors=5,metric='minkowski',p=1)
scores = cross_val_score(knn, x, y, cv = k_folds)
KNN=scores.mean()

print("Cross Validation Scores: ", scores)
print("Average CV Score: ", scores.mean())
```

# **Random Forest classifier**

```python
rfc=RandomForestClassifier()
scores = cross_val_score(rfc, x, y, cv = k_folds)
RF=scores.mean()

print("Cross Validation Scores: ", scores)
print("Average CV Score: ", scores.mean())
```

# **Decision Tree Classifier**

```python
dt=DecisionTreeClassifier(criterion = "entropy",random_state=20)
scores = cross_val_score(dt, x, y, cv = k_folds)
DT=scores.mean()

print("Cross Validation Scores: ", scores)
print("Average CV Score: ", scores.mean())
```

# **Support Vector Machine**

```python
svm = SVC(kernel="rbf", gamma=0.5, C=1.0)
scores = cross_val_score(svm, x, y, cv = k_folds)
SVM=scores.mean()

print("Cross Validation Scores: ", scores)
print("Average CV Score: ", scores.mean())
```

# **Naive Bayes**

```python
gnb = GaussianNB()
scores = cross_val_score(gnb, x, y, cv = k_folds)
NB=scores.mean()

print("Cross Validation Scores: ", scores)
print("Average CV Score: ", scores.mean())
```

# **Cross validation Scores**

```python
Accuracy={"Logistic_Regression_Classifier":LR,"KNN_Classifier":KNN,"Random_Forest_Classifier":RF,"Descion_Tree":DT,"Support_Vector_Machine":SVM,"Naive_Bayes":NB}
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt_1 = plt.figure(figsize=(14, 5))
a=sns.barplot(x=list(Accuracy.keys()),y=list(Accuracy.values()))
for i,v in enumerate(list(Accuracy.values())):
  a.text(i,v+0.0,f'{v:.0%}',ha='center')
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
plt.xlabel("Machine Learning Classification Algorithms",fontdict = font2)
plt.ylabel("Accuracy",fontdict = font2)
plt.title("Machine Learning Classifier Algorithms",fontdict = font1)
plt.show()
```
